```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['sex']=lb.fit_transform(df['sex'])
df['smoker']=lb.fit_transform(df['smoker']) 
df['region']=lb.fit_transform(df['region'])
```

```python
print (df.head(2))
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
df_covid=pd.read_csv("covid_toy.csv")
```

```python
df_covid.head()
```

```python
df_covid.isnull().sum()
```

```python
# to fill missing values
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer()
```

```python
df_covid['fever']=si.fit_transform(df_covid[['fever']])
```

```python
df_covid.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df_covid['gender']=lb.fit_transform(df_covid['gender'])
df_covid['cough']=lb.fit_transform(df_covid['cough'])
df_covid['city']=lb.fit_transform(df_covid['city'])
df_covid['has_covid']=lb.fit_transform(df_covid['has_covid'])
```

```python
df_covid.head()
```

```python
x=df_covid.drop(columns=['has_covid'])
y=df_covid['has_covid']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
