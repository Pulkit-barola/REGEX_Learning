```python
df_covid=pd.read_csv("covid_toy.csv")
```

```python
df_covid.head()
```

```python
df_covid.isnull().sum()
```

```python
# to fill missing values
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer()
```

```python
df_covid['fever']=si.fit_transform(df_covid[['fever']])
```

```python
df_covid.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df_covid['gender']=lb.fit_transform(df_covid['gender'])
df_covid['cough']=lb.fit_transform(df_covid['cough'])
df_covid['city']=lb.fit_transform(df_covid['city'])
df_covid['has_covid']=lb.fit_transform(df_covid['has_covid'])
```

```python
df_covid.head()
```

```python
x=df_covid.drop(columns=['has_covid'])
y=df_covid['has_covid']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# TP (TRUE POSITIVE) ----> HUMAN (FEVER) & MACHINE (FEVER) => GIVE RIGHT PREDICTION
# TN (TRUE NEGATIVE) ----> HUMAN (FEVER) & MACHINE (FEVER) => GIVE RIGHT PREDICTION
# FP (FALSE POSITIVE) ---->HUMAN (FEVER) & MACHINE (FEVER) => GIVE RIGHT PREDICTION
# FN (FALSE NEGATIVE) ---->HUMAN (FEVER) & MACHINE (FEVER) => GIVE RIGHT PREDICTION
```

```python
# hyperparameter tunnig -> to make the accuracy high

# hyperparameter tuning is used to improve model performance, including increasing accuracy (or other metrics like precision, recall, F1, depending on the task).

```

```python
# Example: Decision Tree
# If you're using a DecisionTreeClassifier, some key hyperparameters are:

# max_depth: How deep the tree can go

# min_samples_split: Minimum samples required to split a node

# criterion: Function to measure quality of a split (e.g., "gini", "entropy")

# If max_depth is too high → the model overfits (good on train data, bad on test).
# If it's too low → it underfits (too simple to learn patterns).

# By tuning these hyperparameters, you can find the sweet spot that gives better accuracy on validation/test data.
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}
```

```python
grid = GridSearchCV(DecisionTreeClassifier(),param_grid, cv=5)
grid.fit(x_train, y_train)
```

```python
print("Best accuracy:", grid.best_score_)
print("Best params:", grid.best_params_)
```
